## Loading Necessary Libararies

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#Resizing the graph(dpi specifies pixels per inch. when saving specify dpi as 300)
plt.figure(figsize=(3,3),dpi=100)

#Line 1

# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')


#fmt=[color][marker][line]
plt.plot(x,y,'b.-',label='2x')

#Line 2

#select intervals we want plot points at
x2=np.arange(0,4.5,0.5)

#plot part of the graph as a line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

#plot remainder of the graph as a dotted line
plt.plot(x2[5:],x2[5:]**2, 'r--')

#to give title to the graph
#fontdict is used for all the font purposes
plt.title('My first graph!',fontdict={'fontname':'Comic sans MS', 'fontsize':20})

#x and y labels
plt.xlabel('X axis')
plt.ylabel('Y axis')

#to add the x and y ticks
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8])

#to add legend to the graph
plt.legend()

#save the figure
plt.savefig('mygraph.png',dpi=300)
plt.show()
```

### Bar chart

```python
labels=['A','B','C']
values=[1,4,2]

plt.figure(figsize=(5,3),dpi=100)

bars = plt.bar(labels,values)

patterns=['*','o','+']

for bar in bars:
    bar.set_hatch(patterns.pop(0))
    
    
plt.title('Bar Chart', fontdict={'fontname':'Calibri','fontsize':20})

plt.xlabel('Labels')
plt.ylabel('Values')

plt.savefig('BarChart.png',dpi=300)
plt.show()
```
